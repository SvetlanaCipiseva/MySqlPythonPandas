## Pie chart

```python
%matplotlib inline
import matplotlib.pyplot as plt
import mysql.connector
import pandas as pd
sql_pay_query = """ 
SELECT productLine, SUM(quantityInStock) total
FROM products
GROUP BY productLine
""" 
cnx = mysql.connector.connect(user='root', password='***',
                              host='localhost',
                              database='classicmodels')

df = pd.read_sql_query(sql_pay_query,cnx)
```

```python
plt.pie(df.total, labels=df.productLine,autopct='%1.0f%%',pctdistance=0.8, labeldistance=1.1, radius=1.5)
plt.show()
```

```python
plt.savefig('piechart.png')
```
